```python
columns_valid = []
```

Según el dataset de adidas, para las columnas categóricas, encontrar -aquellas columnas que tienen mas de un nivel, y que por lo menos uno de esos niveles represente mas o igual del 98%. Para las columnas que cumplan esa premisa, eliminaran del array de columnas validas 

```python
df_result = pd.DataFrame()
```

```python
adidas.loc[: , ["name"]].shape[0]
```

```python
for c_ in categoriescolumns:
    i = adidas.loc[: , [c_]].drop_duplicates()
    shapelevel = i.shape[0]
    if shapelevel > 1 :
        columns_valid.append(c_)           
```

```python
df_columns_many_levels = adidas.loc[: , columns_valid]
```

Según el dataset de adidas, para las columnas categóricas, encontrar aquellas columnas que tengan entre 3-6 niveles. Al nivel con la frecuencia mas baja lo vamos a remplazar por la palabra Others, una vez ya tenga esa regla aplicarla al dataset original, tal cual como se hizo con el ejercicio hecho en clases.

```python
many_columns = []
```

```python
for c_ in categoriescolumns:
    i = adidas.loc[: , [c_]].drop_duplicates()
    shapelevel = i.shape[0]
    if (shapelevel >= 3) & (shapelevel <= 6)   :
        many_columns.append(c_)           
```

```python
many_columns
```

```python
df = adidas.loc[: , many_columns]
```

```python
df2 = df.loc[: , "category"].value_counts().rename_axis("category").reset_index(name='counts') 
```

```python
df2["percentage"] = df2["counts"] / df.shape[0] 
```

```python
df2["rank"] = df2["percentage"].rank(ascending = True)
```

```python
df2
```

```python
df2["category_n"] = df2.apply(lambda r: "Others" if r["rank"] == 1.0 else r["category"], axis = 1)
```

```python
df2 = df2.loc[: , ["category", "counts", "percentage","category_n"]]
```

```python
adidas = adidas.merge(df2, on = ["category"], how="left")
```
